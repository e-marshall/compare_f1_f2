```python
import xarray as xr
import os 
import numpy as np
```

```python
f1_root_path = "/Users/emmamarshall/Desktop/facts_work/facts_v1/facts/experiments/larmip.ssp245/output/" 
f1_files = [f for f in os.listdir(f1_root_path) if f.endswith('.nc')]
f1_datasets = dict(zip(f1_files, [xr.open_dataset(os.path.join(f1_root_path, f)) for f in f1_files]))
```

```python
f2_root_path = "/Users/emmamarshall/Desktop/facts_work/facts_v2/larmip-ais/data/output" 
f2_files = [f for f in os.listdir(f2_root_path) if f.endswith('.nc')]
f2_datasets = dict(zip(f2_files, [xr.open_dataset(os.path.join(f2_root_path, f)) for f in f2_files]))
```
